```python
from qiskit.quantum_info import Operator
from math import sqrt

X = Operator([[0,1], [1,0]])
Y = Operator([[0,-1j], [1j, 0]])
Z = Operator([[1, 0], [0, -1]])
H = Operator([[1/sqrt(2), 1/sqrt(2)], [1/sqrt(2), -1/sqrt(2)]])
S = Operator([[1, 0], [0, 1j]])
T = Operator([[1, 0], [0, (1+1j)/sqrt(2)]])
display(X.draw("latex"))
display(Y.draw("latex"))
display(Z.draw("latex"))
display(H.draw("latex"))
display(S.draw("latex"))
display(T.draw("latex"))
```

```python
# apply unitary operation

from qiskit.quantum_info import Statevector

ket0 = Statevector([1, 0])
display(ket0.draw("latex"))
ket1 = Statevector([0, 1])
display(ket1.draw("latex"))
# Hadamard
ket_0H = ket0.evolve(H) # H . ket0
display(ket_0H.draw("latex"))

ket_1H = ket1.evolve(H) # H . ket1
display(ket_1H.draw("latex"))

```
