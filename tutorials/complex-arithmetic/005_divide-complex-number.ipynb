#### Divide complex number
**Inputs**:

1. A complex number *x=a+bi*

2. A complex number *y=c+di*

**Goal**: Return the result of the division *$\frac{x}{y}$*. If *y=0* return 0

```python
def divide_complex_number(x: complex, y: complex)->complex:
  try:
    return x / y
  except ZeroDivisionError:
    return "Error: Division by zero is not allowed."

print(divide_complex_number(10+2j, 5+8j))
print(divide_complex_number(10-2j, 0))
```
