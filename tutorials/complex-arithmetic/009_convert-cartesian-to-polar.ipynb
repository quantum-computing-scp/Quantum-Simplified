**Input**: A complex number *x=a+bi*

**Goal**: Return the polar representation *x=r.$e^{i\theta}$*

*r* should be non-negative: *r>=0*

*${\theta}$* should be between -*${\pi}$* and *${\pi}$*: -*${\pi}$* <= *${\theta}$* <= *${\pi}$*

```python
import math
def cartesian_to_polar1 (x:complex):
  r=math.sqrt(x.real**2 + x.imag**2)
  theta=math.atan2(x.imag, x.real)
  print(f"r={r}")
  print(f"Theta={theta}")
  
cartesian_to_polar1(1+1j)
```

```python
import cmath
def cartesian_to_polar2 (x:complex):
  r,theta=cmath.polar(x)
  print(f"r={r}")
  print(f"Theta={theta}")

cartesian_to_polar2(1+1j)
```

```python
import cmath
import math
def polar_to_cartesian(r: float, theta:float)->complex:
  return cmath.rect(r, theta)

print(polar_to_cartesian(math.sqrt(2), math.pi/4))
```
