````python
def imaginary_power(n : int) -> int:
    # If n is divisible by 4
    if n % 4 == 0:
        return 1
    else:
        return 1j ** (n % 4)
x=imaginary_power(4)
y=imaginary_power(7)
print(x)
print(y)
````
