#### Powers of Imaginary Unit
**Input**: An even integer *n*. The integer can be zero, positive or negative, but it is guaranteed to be even

**Goal**: Return *$i^n$*, that is, the *n*-th power of *i*

---
**Solution**: Notice the raising *i* to the power of 4 gives, when the exponent *n* is divisible by 4, *${i^4}$* is always be 1. Fow an even exponent *n* that isn't divisible by 4, you'll have *${i^n}$* is always be -1

*${i^4}$=${i^2}$.${i^2}$=-1.-1=1*

*${i^6}$=${i^2}$.${i^2}$.${i^2}$=-1.-1.-1=-1*

```python
def imaginary_power(n : int) -> int:
    if n % 4 == 0:
        return 1
    else:
        return -1
x=imaginary_power(4)
y=imaginary_power(6)
print(x)
print(y)
```
