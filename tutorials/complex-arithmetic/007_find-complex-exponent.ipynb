### Find complex exponent

**Input**: A complex number *x=a+bi*

**Goal**: Return the complex number **$e^x$**

**Solution**

*$e^x$=$e^{a+bi}$=$e^a$.$e^{bi}$=$e^a$.(cos(b)+i.sin(b))=$e^a$.cos(b)+i.$e^a$.sin(b)*

```python
# math
import math
def complex_exponent1(x:complex)-> complex:
  E = math.e
  Re_new = (E**x.real)*math.cos(x.imag)
  Im_new = (E**x.real)*math.sin(x.imag)

  return complex(Re_new, Im_new)

print(complex_exponent1(10+30j))
```

```python
# cmath
import cmath
def complex_exponent2(x:complex)->complex:
  return cmath.exp(x)

print(complex_exponent1(10+30j))
```
