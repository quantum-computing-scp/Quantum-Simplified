### Find complex power of real number

**Input**:

1. A non-negative number *r*

2. A complex number *x=a+bi*

**Goal**: Return complex number *$r^x$*

**Solution**

Use the property of logarithms to convert the base to *e* : *r=$e^{ln(r)}$* (r>0)

*$r^x$=$e^{ln(r).(a+bi)}$=$e^{ln(r).a+ln(r).bi}$=$e^{ln(r).a}$.$e^{ln(r).bi}$=$e^{ln(r).a}$ . { cos[ln(r).b] + sin[ln(r).b]i }*

```python
import math
def complex_power_real_number(r: float, x: complex) -> complex:
  if(r < 0): return "A negative number"
  E = math.e
  Ln_r = math.log(r, E)
  phase = Ln_r*x.imag
  magnitude = E**(Ln_r*x.real)
  
  Re_new = magnitude*math.cos(phase)
  Im_new = magnitude*math.sin(phase)
  return complex(Re_new, Im_new)

print(complex_power_real_number(10, 2+3j))
print(complex_power_real_number(-10, 2+3j))
```
