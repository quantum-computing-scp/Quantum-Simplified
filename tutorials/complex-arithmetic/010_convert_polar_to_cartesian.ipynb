**Input**: A complex number in polar form *x=r.$e^{i\theta}$*

**Goal**: Return the cartesian representation *x=a+bi*

*r* should be non-negative: *r>=0*

*${\theta}$* should be between -*${\pi}$* and *${\pi}$*: -*${\pi}$* <= *${\theta}$* <= *${\pi}$*

```python
import math
def polar_to_cartesian1(r: float, theta:float)->complex:
  a=r*math.cos(theta)
  b=r*math.sin(theta)
  return complex(a,b)

print(polar_to_cartesian1(math.sqrt(2), math.pi/4))
```

```python
import cmath
import math
def polar_to_cartesian2(r: float, theta:float)->complex:
  return cmath.rect(r, theta)

print(polar_to_cartesian2(math.sqrt(2), math.pi/4))
```
