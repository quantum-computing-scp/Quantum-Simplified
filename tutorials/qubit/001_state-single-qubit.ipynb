**Input**: A qubit in an unknown state $\ket{\psi}$ = $\alpha\ket{0}$ + $\beta\ket{0}$. The amplitudes $\alpha$ and $\beta$ will be real and non-negative

**Goal**: A tuple of two number ($\alpha'$, $\beta'$) - your estimates of the amplitudes $\alpha$ and $\beta$. The absolute errors $|\alpha -\alpha'|$ and $|\beta -\beta'|$ should be less than or equal to 0.001
