**Input**: A complex matrix

**Goal**: Return conjugate of the following matrix

```python
import numpy as np
def conjugate_matrix(matrix):
  np_matrix = np.array(matrix)
  return np_matrix.conjugate()

print(conjugate_matrix([[1+4j, 2-5j],[8,-2j],[4+3j, 1]]))
print(conjugate_matrix([[1+5j, 2],[3-6j, 4j]]))
```
