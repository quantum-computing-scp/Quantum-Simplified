**Input**: 
1. A number

2. A matrix

**Goal**: Return the result of multiplying the following matrix by the scalar

```python
import numpy as np
def scalar_multiplication(r: float, arr):
  matrix=np.array(arr)
  return r*matrix

print(scalar_multiplication(10, [[1,3,3,5],[2,3,6,3],[523,32,2,34]]))
```
