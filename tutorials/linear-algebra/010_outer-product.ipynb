**Input**: Two vectors

**Goal**: Return the outer product of the following vectors

```python
import numpy as np
def outer_product(vector_A, vector_B):
  np_vector_A = np.array(vector_A)
  np_vector_B = np.array(vector_B)
  transpose_conjugate_vector_B = np_vector_B.T.conjugate()
  return np_vector_A@transpose_conjugate_vector_B

print(outer_product([[-3j],[9]], [[9j],[2]]))
```
