**Input**: Two matrixes

**Goal**: Return the product of the following two matrixes

```python
import numpy as np
def multiply_two_matrix(matrix_A, matrix_B):
  try:
    A=np.array(matrix_A)
    B=np.array(matrix_B)
    return A@B
  except ValueError as e:
    return f"Have an error {e}"

print(multiply_two_matrix([[1,2,3],[4,5,6]],[[1],[2],[3]]))
print(multiply_two_matrix([[5,0,8],[4,3,9],[2,1,7]],[[1,8],[4,6],[2,7]]))
print(multiply_two_matrix([[1,2,3],[4,5,6]],[[1],[2]]))
```
