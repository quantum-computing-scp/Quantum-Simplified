**Input**: Two matrixes

**Goal**: Return the tensor product of the following matrixes

```python
import numpy as np
def tensor_product(matrix_A, matrix_B):
  np_matrix_A = np.array(matrix_A)
  np_matrix_B = np.array(matrix_B)
  
  return np.kron(np_matrix_A, np_matrix_B)

print(tensor_product([[1,2],[3,4]], [[5,6],[7,8]]))
print(tensor_product([[1,2j],[3,4+3j]], [[5,6],[7,8]]))
```
