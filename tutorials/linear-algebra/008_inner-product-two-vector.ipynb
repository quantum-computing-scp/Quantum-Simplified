**Input**: Two complex vector V, W

**Goal**: Return the inner product $\langle V, W \rangle$ of the following vectors

```python
import numpy as np
def inner_product(matrixV, matrixW):
  np_matrixV = np.array(matrixV)
  np_matrixW = np.array(matrixW)
  np_matrixV_conjugate_transpose = np_matrixV.T.conjugate()
  return np_matrixV_conjugate_transpose@np_matrixW

print(inner_product([[-6],[9j]],[[3],[-8]]))
```
