**Input**: A vector

**Goal**: Return the normalized vector V, defined $\frac{V}{||V||}$

```python
import numpy as np
def normalized_vector(vector):
  matrix = np.array(vector)
  conjugate_transpose_matrix = matrix.T.conjugate()
  norm_vector = np.sqrt((conjugate_transpose_matrix@matrix).item())
  return matrix * 1/norm_vector

print(normalized_vector([[-6],[8j]]))
```
