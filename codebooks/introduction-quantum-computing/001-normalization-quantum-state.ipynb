#### Input: You are given an unnormalized vector

### Output: Return normalize vector

```python
import numpy as np
from qiskit.visualization import array_to_latex

ket_0 = np.array([1, 0])
ket_1 = np.array([0, 1])

def normalize_state(alpha, beta):
  norm = np.abs(alpha) ** 2 + np.abs(beta) ** 2
  alphaNormalize = alpha/np.sqrt(norm)
  betaNormalize = beta/np.sqrt(norm)
  return alphaNormalize*ket_0 + betaNormalize*ket_1

print(normalize_state(3+1j,4-2j))
display(array_to_latex(normalize_state(3+1j,4-2j)))
  
```
