```python
import numpy as np

U = np.array([[1, 1], [1, -1]])/np.sqrt(2)

def initialize_qubit():
  return np.array([1, 0])

def apply_U(state):
  return U@state

def measure_state(state, num_meas):
  P_0 = np.abs(state[0])**2
  P_1 = np.abs(state[1])**2
  sample = np.random.choice([0,1], num_meas, True, [P_0, P_1])
  return sample

def quantum_algorithm():
  state = initialize_qubit()
  state_U = apply_U(state)
  return measure_state(state_U, 100)

print(quantum_algorithm())
```
